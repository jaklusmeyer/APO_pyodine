# Clean observations from the I2

## Background

With the I2-cell method, we have a lot of observations, all of which contain the I2 lines along with the stellar spectrum - while that is good for computing RVs, they are in a way "polluted" from the perspective of other potential analyses. So it would be nice if there was a way to "clean" these observations - and there is, as [Diaz et al. (2019)](https://ui.adsabs.harvard.edu/abs/2019AJ....157..204D/abstract) have shown. Their approach is implemented in **pyodine**, and we will demonstrate it here.

The basic idea is: If we have successfully [modelled an observation](./observation.ipynb), the differences between our model $\hat{I}_*(x)$ and the observed spectrum $I_*(x)$ should be some combination of noise $N(x)$ and changes in the stellar spectrum as compared to the input stellar template ($T_*(x)$). All we need to do is convolve the stellar template with the LSF model $L(x)$:

$$
    \hat{T}_*(x) = T_*(x) * L(x)
$$

To this LSF-affected stellar template $\hat{T}_*(x)$ we now add the residuals between observed spectrum (normalized by the continuum model $k(x)$) and our model of it, $\big( \frac{I_*(x)}{k(x)} - \hat{I}_*(x) \big)$:

$$
    \hat{I}_{*}^\mathrm{I2-free}(x) = \hat{T}_*(x) + \big( \frac{I_*(x)}{k(x)} - \hat{I}_*(x) \big)
$$

A few things to note:

* To run the code below, you will need results from the normal **pyodine** modelling of each observation - we make use here of the already analyzed tutorial data, so you will have to first run the tutorial steps to [create a deconvolved stellar template](./template.ipynb) and [model the observations](./observation.ipynb). We will also use the final weighted RVs of each observation instead of the individual chunk best-fit velocities in an attempt to create more uniform chunk models, so also the [velocity combination](./velocities.ipynb) needs to have been performed.

* The employed algorithm includes two different ways to treat the LSF: The first is to simply use the best-fit model LSF of each individual chunk when cleaning that chunk spectrum. This can of course give bad results for chunks where the fitting due to some reason did not go optimally, and it can generally introduce large differences between chunks. So the second option is to smooth the model LSF over neighbouring chunks, which should also be a better representation of instrumental reality (the LSF should not change very abruptly between segments of few dozen pixels). We will do that here, and hand a dictionary to the function below, which specifies over how many orders/pixels to smooth the LSF (and how many pixels the orders are roughly separated by in the raw spectrum).

* Of course, what you can do with the results from this treatment strongly depends on the S/N of the observation AND of the stellar template observation. If these two are not great, the noise will dominate in the estimate of the I2-free observations. [Diaz et al. (2019)](https://ui.adsabs.harvard.edu/abs/2019AJ....157..204D/abstract) achieved an accuracy of 1 to 1.5% in their analysis.

* The results from this algorithm are flux, wavelength and continuum values for each chunk, and they are saved as HDF5-files. This means, they are not yet stitched together to orders or even full spectra - this is something you will still have to implement yourself if needed! Also, the chunk spectra are still convolved with the spectrograph LSF! If you want to receive LSF-free spectra, you will have to apply a deconvolution algorithm of the kind as used in the creation of the devonvolved stellar templates.

## Run the code

Again, first let's set up the path and import the required packages (don't forget: change the pathname when running it on your machine):

```python
# Automatic reloading of imports
%load_ext autoreload
%autoreload 2

import sys
import os
import matplotlib.pyplot as plt
import numpy as np
from astropy.constants import c

# Put in the pyodine path on your machine here!
pyodine_path = '/home/paul/pyodine/'

sys.path.append(pyodine_path)

import pyodine
import pyodine_clean_i2       # <- the I2 cleaning routines
```

And, as always, import the appropriate utilities module and parameter input object:

```python
import utilities_song as utilities

Pars = utilities.pyodine_parameters.Parameters()
```

Now we define the input and output paths. In goes the result from the second modelling run (i.e. with MultiGaussian LSF model) of the first tutorial observation. The output is saved as HDF5-file. We create a plot directory where the results are saved in, but note that in the current implementation of this code no actual plots are created (this may still change).

```python
# Observation name to analyze
obs_name = 's1_2015-05-02T03-48-48_ext'

# Observation result to use (the first one)
res_file = os.path.join(pyodine_path, 'tutorial_data/obs_results/{}/{}_res1.h5'.format(
                        obs_name, obs_name))

# Plot directory
plot_dir = os.path.join(pyodine_path, 'tutorial_data/i2_free/' + obs_name)

# The output filename for the I2-free observation
out_file = os.path.join(plot_dir, obs_name + '_res1_I2free.h5')

# Log files
error_file = os.path.join(plot_dir, 'error.log')
info_file = os.path.join(plot_dir, 'info.log')
```

Also, we want to load the velocity shift of that observation, to also give it as input. Otherwise, the routine would use the respective best-fit chunk velocity for each chunk - but really, all should have the same shift (also we know some chunks perform bad, so want to preven that).

```python
# The filepath of the combined velocity product
comb_vel_file = os.path.join(pyodine_path, 
                             'tutorial_data/vel_results/sigdra_tutorial_comb.h5')

# Load the combined results
comb_vels = pyodine.timeseries.base.CombinedResults(comb_vel_file)

# Find the index of our observation
ind = [i for i in range(len(comb_vels.orig_filename)) if obs_name in comb_vels.orig_filename[i]]

# And get the respective RV (NOT barycentric-corrected!!!)
rv = comb_vels.rv[ind[0]]
```

Finally, we specify the LSF smoothing dictionary, as discussed in the beginning (the parameters should be ok for SONG spectra):

```python
lsf_dict = {
    'smooth_pixels': 160,
    'smooth_orders': 3,
    'order_separation': 15
}
```

And with that we run the I2 cleaning function for a single observation input.

```python
pyodine_clean_i2.clean_i2_single_observation(
    utilities, Pars, res_file, out_file, rv=rv,
    plot_dir=plot_dir, error_log=error_file, 
    info_log=info_file, use_progressbar=True,
    smooth_lsf=True, lsf_dict=lsf_dict)
```

Ok great, the code restored the observation chunks and fit results from the results file, loaded the stellar template and I2 atlas, and performed the cleaning algorithms explained in the top of this section. As you notice, it's quite quick, because no actual modelling is performed.

## Inspecting the result

To inspect the results, we load the saved I2-free data from the HDF5-file. The code block below shows you what kind of information is stored there.

```python
import h5py

with h5py.File(out_file, 'r') as h:
    i2free = pyodine.lib.h5quick.h5data(h)

print('Result keys:')
print(i2free.keys())

print('\n"info" keys:')
print(i2free['info'].keys())

print('\nShape of spectrum results (nr_chunks, nr_pixels):')
print(i2free['flux'].shape)

wave = i2free['wave']
flux = i2free['flux']
cont = i2free['cont']
```

Now we want to compare the cleaned chunk spectra with the original observation, the I2 template spectrum and the deconvolved stellar template, so lets restore/load the respective data.

```python
# Load original result of observation
obs_chunks, fit_results = pyodine.fitters.results_io.restore_results_object(
    utilities, res_file)

# And the deconvolved stellar template
temp_spec = fit_results[0].model.stellar_template

# And the I2 template spectrum
iod_spec = fit_results[0].model.iodine_atlas
```

Alright, let's plot the I2-free spectrum, normalized original observation and I2 template spectrum of one chunk.

```python
# Which chunk to plot
plot_chunk = 170

# Corresponding I2 region
iod_ind = np.where((iod_spec.wave > np.min(wave[plot_chunk])) & (iod_spec.wave < np.max(wave[plot_chunk])))

fig = plt.figure(figsize=(10,6))

plt.plot(wave[plot_chunk], flux[plot_chunk], label='I2-free observation')
plt.plot(obs_chunks[plot_chunk].wave, obs_chunks[plot_chunk].flux/cont[plot_chunk],
         label='Observation (normalized)')
plt.plot(iod_spec.wave[iod_ind], iod_spec.flux[iod_ind], alpha=0.5, label='I2 atlas')

plt.legend()
plt.xlabel(r'Wavelength [$\AA$]')
plt.show()
```

As you can see the I2-free spectrum shows much less absorption than the original observation - obviously, that is due to the missing I2 contribution. You will also notice that the normalization of the observation spectrum has not worked perfectly - one would still have to scale that down by some factor to have its continuum correspond to the I2 and I2-free continuums.

Next, we want to compare the I2-free spectrum to the deconvolved stellar template. Here, we need to take the Doppler shift of the observation relative to the template into account (based on the `rv` value from above). Also there are again issues with non-aligned continuum, so we include a continuum correction factor (just roughly guessing it). We also plot the residuals between I2-free spectrum and stellar template, and for that we need to rebin the latter to the wavelength vector of the observation.

```python
# Which chunk to plot
plot_chunk = 170

# Continuum correction
cont_corr = 0.93

# The Doppler shift between template and observation
z = rv / c.value

# Corresponding template region
temp_ind = np.where((temp_spec[plot_chunk].wave > np.min(wave[plot_chunk])) &
                   (temp_spec[plot_chunk].wave < np.max(wave[plot_chunk])))
temp_wave = temp_spec[plot_chunk].wave[temp_ind]
temp_flux = temp_spec[plot_chunk].flux[temp_ind]

# Rebin the template to the observation wavelengths
# and compute the residuals between both
temp_rebin = pyodine.lib.misc.rebin(temp_wave, temp_flux,
                                    wave[plot_chunk] * (1-z))
residuals = flux[plot_chunk] * cont_corr - temp_rebin

fig = plt.figure(figsize=(10,8))
gs = fig.add_gridspec(2, 1,  height_ratios=(1, 0.4))

ax1 = fig.add_subplot(gs[0])
ax1.plot(wave[plot_chunk] * (1-z), flux[plot_chunk] * cont_corr, 
         label='I2-free observation (rv = {:.1f} km/s)'.format(rv/1000.))
ax1.plot(temp_wave, temp_flux, label='Deconvolved template')
ax1.legend()
ax1.tick_params(axis='both', top=True, right=True, direction='in', which='both', labelbottom=False)

ax2 = fig.add_subplot(gs[1])
ax2.plot(wave[plot_chunk] * (1-z), residuals, 
         label='Residuals (rms = {:.1f}%)'.format(np.std(residuals/temp_rebin)*1e2))
ax2.legend()
ax2.tick_params(axis='both', top=True, right=True, direction='in', which='both')
ax2.set_xlabel(r'Wavelength [$\AA$]')
fig.subplots_adjust(hspace=0)
plt.show()
```

The stellar template has narrower and deeper absorption lines, which is due to the fact that it is LSF-free, while the I2-free spectrum is the stellar spectrum "as the spectrograph would see it", i.e. convolved with the LSF. Also you can notice that the latter is not quite as smooth in the continuum regions as the stellar template - this is the influence of noise from the observation.

Finally, are you interested what the smoothed model LSF looks like? With the following code we can build it and take a look... (this is basically taken from the function `clean_i2_single_observation`)

```python
# The red. Chi^2 serves as weights for the contribution of
# different chunks
redchi2 = np.array([r.redchi for r in fit_results])

# Compute the smoothed evaluated LSFs for each chunk
lsf_smoothed = pyodine.lib.misc.smooth_lsf(
        obs_chunks, lsf_dict['smooth_pixels'], lsf_dict['smooth_orders'], 
        lsf_dict['order_separation'], fit_results,
        redchi2=redchi2, osample=fit_results[0].model.osample_factor)
```

Now we plot the smoothed LSF of a chunk, and also evaluate and plot the original best-fit model LSF of that chunk for comparison.

```python
# Which chunk to plot
plot_chunk = 170

# Create the non-smoothed chunk LSF
x_lsf, lsf_chunk = fit_results[plot_chunk].model.eval_lsf(fit_results[plot_chunk].params)

fig = plt.figure(figsize=(10,6))

plt.plot(x_lsf, lsf_smoothed[plot_chunk], alpha=0.7, label='Smoothed LSF')
plt.plot(x_lsf, lsf_chunk, alpha=0.7, label='Original LSF')
plt.legend()
plt.xlabel('Pixel')
plt.show()
```

You'll notice slight differences between the two, particularly on the left wing. And here is a more extreme chunk (in the red part of the spectrum, i.e. higher chunk numbers, the combination of tellurics and reduced density of I2 lines can often lead to weird LSF shapes).

```python
# Which chunk to plot
plot_chunk = 461

# Create the non-smoothed chunk LSF
x_lsf, lsf_chunk = fit_results[plot_chunk].model.eval_lsf(fit_results[plot_chunk].params)

fig = plt.figure(figsize=(10,6))

plt.plot(x_lsf, lsf_smoothed[plot_chunk], alpha=0.7, label='Smoothed LSF')
plt.plot(x_lsf, lsf_chunk, alpha=0.7, label='Original LSF')
plt.legend()
plt.xlabel('Pixel')
plt.show()
```

```python
import dill
```

```python
class A:
    
    dic = {
        'a': 1,
        'b': 2
    }
    
    @classmethod
    def change(cls, dic):
        cls.dic = dic
```

```python
a = A
a.dic
```

```python
a.change({'a': 4, 'b': 5})
a.dic
```

```python
b = A
b.dic
```

```python
with open('save.pkl', 'wb') as f:
    dill.dump(a, f)
```

```python
with open('save.pkl', 'rb') as f:
    c = dill.load(f)
```

```python
c.dic
```
